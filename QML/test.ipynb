```python
import numpy as np
import pickle
from qiskit.circuit.library import ZZFeatureMap, RealAmplitudes
from qiskit_machine_learning.neural_networks import SamplerQNN
from qiskit_aer.primitives import Sampler as AerSampler
```

```python
loaded_weights = np.load("qml_trained_weights.npy")
```

```python
feature_map = ZZFeatureMap(feature_dimension=3, reps=2)
ansatz = RealAmplitudes(num_qubits=3, reps=4)
sampler = AerSampler()
circuit = feature_map.compose(ansatz)

def interpret(x):
    return x % 3

qnn = SamplerQNN(
    circuit=circuit,
    sampler=sampler,
    input_params=feature_map.parameters,
    weight_params=ansatz.parameters,
    interpret=interpret,
    output_shape=3
)
```

```python
with open("scaler.pkl", "rb") as f:
    scaler = pickle.load(f)
```

```python
new_sample = np.array([[0, 0.0, 0.0]])
new_sample_scaled = scaler.transform(new_sample)
```

```python
output = qnn.forward(new_sample_scaled, loaded_weights)
print(output)
prediction = np.argmax(output)
print("Predicted Label:", prediction)
```
